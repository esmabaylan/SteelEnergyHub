```python
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
conn = psycopg2.connect(
    host='timescaledb',
    port=5432,
    database='energydb',
    user='data_writer',
    password='sifre123',
    options='-c search_path=main_data'
)

df = pd.read_sql("""
    SELECT *
    FROM main_data.energy_readings
    ORDER BY time
""", conn)
conn.close()

print(f'Loaded rows: {len(df)}')
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])


df['hour']        = df['time'].dt.hour
df['day']         = df['time'].dt.day
df['month']       = df['time'].dt.month
df['day_of_year'] = df['time'].dt.dayofyear
df['quarter']     = df['time'].dt.quarter
df['is_weekend']  = df['time'].dt.weekday >= 5


def get_shift(hour):
    if 0 <= hour < 8:
        return 'night_shift'   
    elif 8 <= hour < 16:
        return 'day_shift'      
    else:
        return 'evening_shift'  

df['shift'] = df['hour'].apply(get_shift)


def get_tariff_period(hour):
    if 22 <= hour or hour < 6:
        return 'night'   
    else:
        return 'day'    

df['tariff_period'] = df['hour'].apply(get_tariff_period)

def get_day_period(hour):
    if 0 <= hour < 6:
        return 'midnight'
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    else:
        return 'evening'

df['day_period'] = df['hour'].apply(get_day_period)

print('created features:')
print(df[['time', 'hour', 'month', 'shift', 'tariff_period', 'day_period']].head(10))
```

```python
df = df.sort_values('time')
df['usage_ma_1h']  = df['usage_kwh'].rolling(window=4).mean()   
df['usage_ma_4h']  = df['usage_kwh'].rolling(window=16).mean()  
df['usage_ma_24h'] = df['usage_kwh'].rolling(window=96).mean()  


df['usage_diff']     = df['usage_kwh'].diff()
df['usage_pct_change'] = df['usage_kwh'].pct_change() * 100


df['cumulative_daily'] = df.groupby(df['time'].dt.date)['usage_kwh'].cumsum()

print('created rolling features:')
df[['time', 'usage_kwh', 'usage_ma_1h', 'usage_ma_4h', 'usage_diff']].head(20)
```

```python
df['pf_efficiency'] = (df['lagging_pf'] + df['leading_pf']) / 2


df['reactive_power_balance'] = df['lagging_reactive_power'] - df['leading_reactive_power']

df['low_pf_flag'] = df['lagging_pf'] < 80

print('created power factor features:')
df[['time', 'lagging_pf', 'leading_pf', 'pf_efficiency', 
    'reactive_power_balance', 'low_pf_flag']].head(10)
```

```python

mean = df['usage_kwh'].mean()
std  = df['usage_kwh'].std()
df['z_score'] = (df['usage_kwh'] - mean) / std


Q1 = df['usage_kwh'].quantile(0.25)
Q3 = df['usage_kwh'].quantile(0.75)
IQR = Q3 - Q1
df['iqr_outlier'] = (
    (df['usage_kwh'] < Q1 - 1.5 * IQR) |
    (df['usage_kwh'] > Q3 + 1.5 * IQR)
)

df['z_anomaly'] = df['z_score'].abs() > 3

print(f'Z-score anomaly count  : {df["z_anomaly"].sum()}')
print(f'IQR outlier count : {df["iqr_outlier"].sum()}')
df[df['z_anomaly']][['time', 'usage_kwh', 'z_score', 'load_type']].head(10)
```

```python

NIGHT_TARIFF = 1.80  # TL/kWh
DAY_TARIFF   = 3.20  # TL/kWh

df['cost_tl'] = df['tariff_period'].apply(
    lambda x: NIGHT_TARIFF if x == 'night' else DAY_TARIFF
) * df['usage_kwh']


df['cost_if_night'] = df['usage_kwh'] * NIGHT_TARIFF
df['cost_saving']   = df['cost_tl'] - df['cost_if_night']

toplam_maliyet  = df['cost_tl'].sum()
toplam_tasarruf = df[df['tariff_period'] == 'day']['cost_saving'].sum()

print(f'Toplam maliyet          : {toplam_maliyet:,.2f} TL')
print(f'Tasarruf potansiyeli    : {toplam_tasarruf:,.2f} TL')
print(f'Tasarruf oranı          : %{toplam_tasarruf/toplam_maliyet*100:.1f}')
```

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

shift_avg = df.groupby('shift')['usage_kwh'].mean()
shift_avg.plot(kind='bar', ax=axes[0,0], color='#2196F3', edgecolor='white')
axes[0,0].set_title('Shift-wise Average Consumption')
axes[0,0].set_xlabel('Shift')
axes[0,0].set_ylabel('kWh')
axes[0,0].tick_params(axis='x', rotation=0)

tariff_cost = df.groupby('tariff_period')['cost_tl'].sum()
tariff_cost.plot(kind='bar', ax=axes[0,1], 
                  color=['#FF9800', '#2196F3'], edgecolor='white')
axes[0,1].set_title('Cost by Tariff Period (TL)')
axes[0,1].set_xlabel('Tariff Period')
axes[0,1].set_ylabel('TL')
axes[0,1].tick_params(axis='x', rotation=0)

df['pf_efficiency'].hist(bins=30, ax=axes[1,0], 
                          color='#4CAF50', edgecolor='white')
axes[1,0].set_title('Power Factor Efficiency Distribution')
axes[1,0].set_xlabel('Efficiency Score')
axes[1,0].set_ylabel('Frequency')


monthly = df.groupby('month')['usage_kwh'].mean()
monthly.plot(kind='bar', ax=axes[1,1], color='#9C27B0', edgecolor='white')
axes[1,1].set_title('Monthly Average Consumption')
axes[1,1].set_xlabel('Month')
axes[1,1].set_ylabel('kWh')
axes[1,1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python

print('=== NEW FEATURE EXTRACTION SUMMARY ===')
print(f'Total feature count   : {len(df.columns)}')
print(f'Original feature count : 11')
print(f'New derived features  : {len(df.columns) - 11}')
print()
print('New features:')
new_features = ['hour', 'day', 'month', 'day_of_year', 'quarter',
                'is_weekend', 'shift', 'tariff_period', 'day_period',
                'usage_ma_1h', 'usage_ma_4h', 'usage_ma_24h',
                'usage_diff', 'usage_pct_change', 'cumulative_daily',
                'pf_efficiency', 'reactive_power_balance', 'low_pf_flag',
                'z_score', 'iqr_outlier', 'z_anomaly',
                'cost_tl', 'cost_if_night', 'cost_saving']
for f in new_features:
    print(f'  + {f}')

print()
print(f'Anomaly rate  : %{df["z_anomaly"].mean()*100:.3f}')
print(f'Low PF rate : %{df["low_pf_flag"].mean()*100:.1f}')
print(f'Night usage   : %{(df["tariff_period"]=="night").mean()*100:.1f}')
print(f'Day usage : %{(df["tariff_period"]=="day").mean()*100:.1f}')
```

```python
# ================================
# BLOCK 10 — processed_readings tablosuna yaz
# ================================
from psycopg2.extras import execute_batch

# Sadece gerekli kolonları seç
cols = [
    'time', 'usage_kwh', 'lagging_reactive_power', 'leading_reactive_power',
    'co2_tco2', 'lagging_pf', 'leading_pf', 'nsm', 'week_status',
    'day_of_week', 'load_type', 'hour', 'day', 'month', 'quarter',
    'is_weekend', 'shift', 'tariff_period', 'day_period',
    'usage_ma_1h', 'usage_ma_4h', 'usage_diff', 'usage_pct_change',
    'pf_efficiency', 'reactive_power_balance', 'low_pf_flag',
    'z_score', 'z_anomaly', 'cost_tl', 'cost_saving'
]

df_insert = df[cols].dropna(subset=['usage_ma_1h'])  # NaN satırları atla

conn = psycopg2.connect(
    host='timescaledb', port=5432, database='energydb',
    user='data_writer', password='sifre123',
    options='-c search_path=main_data'
)
cur = conn.cursor()

rows = [tuple(row) for row in df_insert.itertuples(index=False)]
placeholders = ','.join(['%s'] * len(cols))

execute_batch(cur, f"""
    INSERT INTO main_data.processed_readings ({','.join(cols)})
    VALUES ({placeholders})
    ON CONFLICT DO NOTHING
""", rows, page_size=1000)

conn.commit()
cur.close()
conn.close()

print(f'Yazılan satır sayısı: {len(rows)}')
```
